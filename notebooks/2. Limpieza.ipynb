Análisis inicial del archivo "sales":

1. Tipo de datos a cambiar: 

    - **SalesDate**: object -> datetime

2. Categorías a cambiar: 

    - Cambiar nombre de las columnas:

        - **SalesID** -> Sales_Id
        - **SalesPersonID** -> Sales_Person_Id
        - **CustomerID** -> Customer_Id
        - **ProductID** -> Product_Id
        - **SalesDate** -> Sales_Date

    - **Sales_Date**: dejar sólo la fecha y crear una nueva columna sólo con la hora (no los minutos) para ver las ventas por horas

    - Crear nueva columna Total_Price obteniendo el precio del producto del archivo 'products' y aplicando el descuento.

3. Columnas con valores nulos: 

    - **SalesDate**. 67526 valores nulos

4. Columnas a eliminar: 

    - **TransactionNumber**: no aporta valor a nuestro análisis
    - **TotalPrice**: todos los valores son ceros

```python
ruta = ru.ruta_raw_data+"/sales.csv" # Ruta del archivo CSV. Cambiar según sea necesario.
df_sales = pd.read_csv(ruta)
# Homogeneizamos los nombres de las columnas.
sa.homogeneizar_nombre_columnas(df_sales)
df_sales.sample(5)
```

```python
# Eliminamos las columnas Total_Price y Transaction_Number.
columnas_a_eliminar = ['Total_Price', 'Transaction_Number']
sa.eliminar_columnas(df_sales, columnas_a_eliminar)
```

```python
# Gestión de nulos de la columna Sales_Date: no los podemos eliminar porque son necesarios para el análisis, así que los rellenamos con la fecha más próxima, usando el método forward fill.
df_sales['Sales_Date'] = df_sales['Sales_Date'].fillna(method='ffill')
```

```python
# Añadimos la columna de sólo hora a partir de la columna Sales_Date.
df_sales['Sales_Hour'] = pd.to_datetime(df_sales['Sales_Date']).dt.hour

# Convertimos la columna Sales_Date a tipo datetime y dejamos solo la fecha.
sa.fecha(df_sales, ['Sales_Date'])
df_sales.sample(5)
```

```python
# Creamos la columna Total_Price a partir de la columna Quantity y Price, y aplicamos el descuento.
df_sales_price = df_sales.merge(df_products[['Product_Id', 'Price']], on='Product_Id', how='left')
df_sales_price['Total_Price'] = round(df_sales_price['Quantity'] * df_sales_price['Price'] * (1 - df_sales_price['Discount']),2)
sa.eliminar_columnas(df_sales_price, ['Price'])
df_sales_price.sample(5)
```

```python
# Comprobamos que todo esté correcto.
sa.preanalisis(df_sales_price)
```

```python
# Guardamos los datos limpios en la carpeta clean_data.
df_sales_price.to_csv(ru.ruta_clean_data+"/sales_clean.csv", index=False) 
```
